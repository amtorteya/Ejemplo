**4.** En el siguiente bloque de código se muestra una gráfica para la <u>población femenina</u>. Para obtener esta parte de la población, se aplica la condición *== "Female"* sobre la columna *Sexo* de los datos. Es decir: <code>data.Sexo == "Female"</code>

Posteriormente, para el **eje Y** se utiliza el nivel de obesidad que se seleccionó en el punto anterior, mientras que en el **eje X** se utiliza el peso, una de las variables cuantitativas de la base de datos. Como sólamente me interesa la población femenina, es necesario utilizar el código descrito anteriormente sobre las dos columnas que se van a utilizar (*nivel de obesidad* y *peso*).

Por último, se utilizó la librería plt para graficar. La función <code>plt.figure</code> ayuda a definir el tamaño del gráfico, y la función <code>plt.scatter</code> grafica los datos indicados.

```python
%matplotlib inline

import matplotlib.pyplot as plt

# Definir "x" y "y", teniendo en cuenta sólo a la población FEMENINA
data_x = data.Peso[data.Sexo == "Female"]
data_y = data.NivelDeObesidad[data.Sexo == "Female"]

# Graficar datos
plt.figure(figsize=(10, 4))
plt.scatter(x = data_x, y = data_y, alpha = 0.05)
plt.show()
```
